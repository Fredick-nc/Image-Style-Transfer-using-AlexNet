## Read records of tensorboard and visualize

```python
from tensorboard.backend.event_processing import event_accumulator
import matplotlib.pyplot as plt
```

```python
def read_tensorboard_data(tensorboard_path, val_name):
    """读取tensorboard数据，
    tensorboard_path是tensorboard数据地址val_name是需要读取的变量名称"""
    ea = event_accumulator.EventAccumulator(tensorboard_path)
    ea.Reload()
    print(ea.scalars.Keys())
    val = ea.scalars.Items(val_name)
    return val
 
def draw_plt(val, val_name):
    """将数据绘制成曲线图，val是数据，val_name是变量名称"""
    plt.figure()
    plt.plot([i.step for i in val], [j.value for j in val], label=val_name)
    """横坐标是step，迭代次数
    纵坐标是变量值"""
    plt.xlabel('step')
    plt.ylabel(val_name)
    plt.show()

if __name__ == "__main__":
    for i in ['summaries_5/total_loss','summaries_5/content_loss','summaries_5/style_loss']:
        tensorboard_path = '/Users/niecheng/Desktop/学习资料/All Subjects Code/毕业设计/Alexnet_Style_Transfer_results/alexnet_graphs/events.out.tfevents.1648174641.改进第十一次.local'
        val = read_tensorboard_data(tensorboard_path, i)
        draw_plt(val, i)
        tensorboard_path1 = '/Users/niecheng/Desktop/学习资料/All Subjects Code/毕业设计/Alexnet_Style_Transfer_results/alexnet_graphs/events.out.tfevents.1648174788.原第十一次.local'
        val1 = read_tensorboard_data(tensorboard_path1, i)
        draw_plt(val1, i)
```
