# 四种深度学习图像处理中常用的非线性激活函数：

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
sns.set(style="darkgrid") 

fig = plt.figure(figsize=(12,6))

plt.xlim([-10, 10]);
plt.ylim([-1, 1.6]);

# 定义数值
x = np.sort(np.linspace(-10,10,1000))

#ReLu 函数
relu= [max(item,0) for item in x]

#ELU函数
alpha = 0.2
elu = [item if item>0 else (np.exp(item)-1)*alpha for item in x]

#Sigmoid 函数
sigmoid = 1/(1+np.exp(-x))

#Tanh 函数
Tanh = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))

# 绘图
plt.plot(x,relu,color="red", label = r"ReLu", marker='*')
plt.plot(x,Tanh,color="#0000ff", label = r"Tanh")
plt.plot(x,elu,color="#00ff00", label = r"ELU",marker='.')
plt.plot(x,sigmoid,color="#00ffee", label = r"Sigmoid")

plt.legend(prop={'family':'Times New Roman','size': 16})
plt.show()
```

```python
from PIL import Image, ImageOps
import numpy as np
import os
def Resize_Image(img_path, height, width, save=True):
    image = Image.open(img_path)     #打开一个图片
    image = ImageOps.fit(image, (width, height), Image.ANTIALIAS)  #返回一个指定大小的裁剪过的图像，过滤器为ANTIALIAS
    if save:
        image_dirs = img_path.split('/')
        image_dirs[-1] = 'resized_' + image_dirs[-1]
        out_path = '/'.join(image_dirs)
        if not os.path.exists(out_path):
            image.save(out_path)
    image = np.asarray(image, np.float32)
    return np.expand_dims(image, 0)
```

```python
# https://blog.csdn.net/zsc201825/article/details/89645190
from cv2 import cv2
import numpy as np
def contrast(img0):   
    img1 = cv2.cvtColor(img0, cv2.COLOR_BGR2GRAY) #彩色转为灰度图片
    m, n = img1.shape
    #图片矩阵向外扩展一个像素
    img1_ext = cv2.copyMakeBorder(img1,1,1,1,1,cv2.BORDER_REPLICATE) / 1.0   # 除以1.0的目的是uint8转为float型，便于后续计算
    rows_ext,cols_ext = img1_ext.shape
    b = 0.0
    for i in range(1,rows_ext-1):
        for j in range(1,cols_ext-1):
            b += ((img1_ext[i,j]-img1_ext[i,j+1])**2 + (img1_ext[i,j]-img1_ext[i,j-1])**2 + 
                    (img1_ext[i,j]-img1_ext[i+1,j])**2 + (img1_ext[i,j]-img1_ext[i-1,j])**2)

    cg = b/(4*(m-2)*(n-2)+3*(2*(m-2)+2*(n-2))+2*4) #对应上面48的计算公式
    print(cg)

img0 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/199_fixed.png')
contrast(img0)
img1 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/199_expdecay.png')
contrast(img1)
img2 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/199纵向对比.png')
contrast(img2)
img3 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/resized_Gatys对比.png')
contrast(img3)
img4 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/Johnson方法.png')
contrast(img4)
img5 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/resized_Huang方法.png')
contrast(img5)
img6 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/199.png')
contrast(img6)
img7 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/alextest.png')  #Alexnet手动搭建生成
contrast(img7)
img8 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/resized_alextest.png')  #resize看来会导致失真
contrast(img8)
img9 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/199_shifted.png')  #gram矩阵偏移量能提升效果
contrast(img9)
img10 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/199_RMS.png')  
contrast(img10)
# img11 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/499.png')  #迭代500次，固定学习率
# contrast(img11)
img12 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/459_RMS最优.png')  #迭代500次，固定学习率
contrast(img12)
img13 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/499_constant.png')  #迭代500次，固定学习率
contrast(img13)
img14 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/2499_改进.png')  #迭代2500次，固定学习率，2499_改进
contrast(img14)
img15 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/2499_原.png')  #迭代2500次，固定学习率，2499_原
contrast(img15)

img16 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_原_1000.png')  #第一次实验
contrast(img16)
img17 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_改进_1000.png')  
contrast(img17)

img18 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_改进第八次.png')  #第二次实验
contrast(img18)
img19 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_原第八次.png')  
contrast(img19)

img20 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_改进第九次.png')  #第三次实验
contrast(img20)
img21 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_原第九次.png')  
contrast(img21)

img22 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_改进第十次.png')  #第四次实验
contrast(img22)
img23 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_原第十次.png')  
contrast(img23)

img24 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_改进第十一次.png')  #第五次实验
contrast(img24)
img25 = cv2.imread('./Alexnet_Style_Transfer_results/alexnet_outputs/999_原第十一次.png')  
contrast(img25)
#结论：图片越清晰，对比度越大
```

## 峰值信噪比：

WIKI解释
峰值信噪比（英语：Peak signal-to-noise ratio，常缩写为PSNR）是一个表示信号最大可能功率和影响它的表示精度的破坏性噪声功率的比值的工程术语。由于许多信号都有非常宽的动态范围，峰值信噪比常用对数分贝单位来表示。

计算PSNR要先知道MSE(均方误差)的计算。两个m×n单色图像I和K，如果一个为另外一个的噪声近似，那么它们的的均方误差定义为：

MSE的概念应该是比较熟悉的，这也是常见的损失函数。而PSNR就是通过MSE得出来的，公式如下：

其中，MAXI是表示图像点颜色的最大数值，如果每个采样点用 8 位表示，那么就是 255。

所以MSE越小，则PSNR越大；所以PSNR越大，代表着图像质量越好。



### 本文方法：

```python
#计算峰值信噪比
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
 
def read_img(path):
    return tf.image.decode_image(tf.compat.v1.read_file(path))
 
def psnr(tf_img1, tf_img2):
    return tf.image.psnr(tf_img1, tf_img2, max_val=255)
 
def _main():
    CONTENT_IMAGE_PATH = './Test_dataset/000000263337.jpg'
    t1 = read_img(CONTENT_IMAGE_PATH)
    t_Alexnet = read_img('./comparison/本文方法/999_改进_1000.png')
    with tf.compat.v1.Session() as sess:
        sess.run(tf.compat.v1.global_variables_initializer())
        y1 = sess.run(psnr(t1, t_Alexnet))
        print(y1)

if __name__ == '__main__':
    _main()
```

```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np 
import math

def BiLinear_interpolation(img,dstH,dstW):
    scrH,scrW,_=img.shape
    img=np.pad(img,((0,1),(0,1),(0,0)),'constant')
    retimg=np.zeros((dstH,dstW,3),dtype=np.uint8)
    for i in range(dstH):
        for j in range(dstW):
            scrx=(i+1)*(scrH/dstH)-1
            scry=(j+1)*(scrW/dstW)-1
            x=math.floor(scrx)
            y=math.floor(scry)
            u=scrx-x
            v=scry-y
            retimg[i,j]=(1-u)*(1-v)*img[x,y]+u*(1-v)*img[x+1,y]+(1-u)*v*img[x,y+1]+u*v*img[x+1,y+1]
    return retimg


im_path='Alexnet_Style_Transfer_results/alexnet_outputs/199_fixed.png'
image=np.array(Image.open(im_path))
image2=BiLinear_interpolation(image,image.shape[0]*2,image.shape[1]*2)
image2=Image.fromarray(image2.astype('uint8')).convert('RGB')
image2.save('Alexnet_Style_Transfer_results/alexnet_outputs/out.png')
```
